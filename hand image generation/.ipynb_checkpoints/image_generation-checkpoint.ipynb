# Helpers

```python

def set_seed(seed):
    """
    Use this to set ALL the random seeds to a fixed value and take out any randomness from cuda kernels
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.benchmark = False  ## uses the inbuilt cudnn auto-tuner to find the fastest convolution algorithms. -
    torch.backends.cudnn.enabled   = False

    return True


set_seed(42)

def weights_init(m):

    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
class HandDataset(Dataset):
    def __init__(self, data_path="real_hands", transformations=None):
        self.data_path = data_path
        self.transformations = transformations
        self.data = self.load_images_to_numpy_array(self.data_path)
            
    def load_images_to_numpy_array(self, directory):

        file_names = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

        # Initialize an empty list to store the images
        images = []

        # Loop over the file names and load images
        for file_name in file_names:
            img_path = os.path.join(directory, file_name)
            try:
                # Open the image without converting to RGB
                img = Image.open(img_path)

                # Convert the image to a NumPy array and append to the list
                images.append(np.array(img))
            except IOError:
                print(f'Error loading image: {img_path}')

        # Convert the list of images to a NumPy array
        return np.array(images, dtype=float)

    def __getitem__(self, idx):
        sample = self.data[idx]
        
        if self.transformations:
            sample = self.transformations(sample)
        
        return sample 
        
    def __len__(self):
        return self.data.shape[0]
```

```python
train_transform = transforms.Compose([
    transforms.ToTensor(),
    # standardisation and normalisation
    transforms.Lambda(lambda x : x / 255.),
    Normalize(mean=[0.5], std=[0.5]),
])

# load data local
hand_data = HandDataset(
    transformations = train_transform
)

# load data colab
# hand_data = HandDataset(
#     transformations = train_transform,
#     data_path=f"{colab_path}/real_hands"
    
# )
```

## Network design and implementation

```python
# https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html

# Root directory for dataset
dataroot = "real_hands"

# Number of workers for dataloader
workers = 1

# Batch size during training
batch_size = 64

# Spatial size of training images. All images will be resized to this
#   size using a transformer.
image_size = 32

# Number of channels in the training images. For color images this is 3
nc = 1

# Size of z latent vector (i.e. size of generator input)
nz = 100

# Size of feature maps in generator
ngf = 64

# Size of feature maps in discriminator
ndf = 64

# Number of training epochs
num_epochs = 5

# Learning rate for optimizers
lr = 0.0008

# Beta1 hyperparameter for Adam optimizers
beta1 = 0.5

# Number of GPUs available. Use 0 for CPU mode.
ngpu = 1
```

```python
import torch.nn as nn

class Generator(nn.Module):
    
    # ===================================================================
    # The structure of my GAN is a DCGAN
    # I followed a pytorch tutorial 
    # https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html
    # ===================================================================
    
    def __init__(self, ngpu, nz, ngf, nc):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.nz = nz
        self.ngf = ngf
        self.nc = nc
        self.relu = nn.ReLU()
        
        # Define individual layers
        self.conv1 = nn.ConvTranspose2d(nz, ngf * 8, 4, 1, 0, bias=False)
        self.batch_norm1 = nn.BatchNorm2d(ngf * 8)

        self.conv2 = nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False)
        self.batch_norm2 = nn.BatchNorm2d(ngf * 4)

        self.conv3 = nn.ConvTranspose2d(ngf * 4, ngf * 2, 4, 2, 1, bias=False)
        self.batch_norm3 = nn.BatchNorm2d(ngf * 2)

        self.conv4 = nn.ConvTranspose2d(ngf * 2, nc, 4, 2, 1, bias=False)
        self.tanh = nn.Tanh()

    def forward(self, x):
        x = x.view(-1, nz, 1, 1)
        x = self.relu(self.batch_norm1(self.conv1(x)))
        x = self.relu(self.batch_norm2(self.conv2(x)))
        x = self.relu(self.batch_norm3(self.conv3(x)))
        x = self.tanh(self.conv4(x))  
        return x

```

```python
import torch.nn as nn

class Discriminator(nn.Module):
    # ===================================================================
    # The structure of my GAN is a DCGAN
    # I followed a pytorch tutorial 
    # https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html
    # ===================================================================
    def __init__(self, ngpu, nc, ndf):
        super(Discriminator, self).__init__()
        
        self.ngpu = ngpu
        self.nc = nc
        self.ndf = ndf
        self.leaky_relu = nn.LeakyReLU(0.2, inplace=True)

        self.conv1 = nn.Conv2d(nc, ndf, 4, 2, 1, bias=False)

        self.conv2 = nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False)
        self.batch_norm2 = nn.BatchNorm2d(ndf * 2)

        self.conv3 = nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False)
        self.batch_norm3 = nn.BatchNorm2d(ndf * 4)

        self.conv4 = nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias=False)
        self.batch_norm4 = nn.BatchNorm2d(ndf * 8)

        self.conv5 = nn.Conv2d(ndf * 8, 1, 2, 1, 0, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.leaky_relu(self.conv1(x))
        x = self.leaky_relu(self.batch_norm2(self.conv2(x)))
        x = self.leaky_relu(self.batch_norm3(self.conv3(x)))
        x = self.leaky_relu(self.batch_norm4(self.conv4(x)))
        x = self.sigmoid(self.conv5(x))
        return x.reshape(-1,1)
```

## Training and evaluation loops

```python
def D_train(x):
    #-------------- Function for the discriminator training -------------------#
    D.train()
    D_optimizer.zero_grad()

    # train discriminator on real data -- assign high score (use 1 here)
    x_real, y_real = x, torch.ones(batch_size, 1)
    x_real, y_real = x_real.to(device), y_real.to(device)

    D_output = D(x_real)
    D_real_loss = criterion(D_output, y_real) ### your code goes here

    # sample vector and produce generator output
    z = torch.randn(batch_size, z_dim).to(device)
    x_fake, y_fake = G(z), torch.zeros(batch_size, 1).to(device)

    D_output = D(x_fake)
    D_fake_loss = criterion(D_output, y_fake)

    # combine the losses
    D_loss = D_real_loss + D_fake_loss
    
    # model update
    D_loss.backward()
    D_optimizer.step()

    return  D_loss.data.item()  ###deprecated version of loss.detach(), basically gets access to the tensor without the computational graph attached
```

```python
def G_train(x):
    
    #-------------- Function for the generator training -------------------#
    G.train()
    G_optimizer.zero_grad()

    # sample vector and produce generator output
    z = torch.randn(batch_size, z_dim).to(device)
    G_output = G(z)

    # obtain scores from D for the generated data
    D_output = D(G_output)

    y = torch.ones(batch_size, 1).to(device)
    G_loss = criterion(D_output, y)

    # model update
    G_loss.backward()
    G_optimizer.step()

    return G_loss.data.item()
```

```python
import torch
from PIL import Image
import os

def save_tensor_images(tensor_array, directory="GAN_hands", file_prefix="acse-jp2923_GAN"):

    """
    Saves a batch of images stored in a torch tensor as JPEG images.

    :param tensor_array: A torch tensor of shape (N, H, W) or (N, H, W, C) representing images
    :param directory: Directory where images will be saved
    :param file_prefix: Prefix for the saved image files
    """
    if not os.path.exists(directory):
        os.makedirs(directory)

    # Normalize the tensor to the range [0, 255] if it's not already
    if tensor_array.max() <= 1:
        tensor_array = tensor_array * 255

    for i, flat_img_tensor in enumerate(tensor_array):
        # Reshape the flattened image tensor to 2D
        img_tensor = flat_img_tensor.view(32,32)
        img = Image.fromarray(img_tensor.byte().cpu().numpy())
        img.save(f"{directory}/{file_prefix}_{i:03d}.jpeg")
        
```
